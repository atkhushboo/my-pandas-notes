```python
import pandas as pd

data={
    "Soap_ID":[101,102,103,104,105,106],
    "Soap_name":[
        "Neem_Soap",
        "Lemon_Soap",
        "Aloavera_Soap",
        "Rose_Soap",
        "Charcoal_Soap",
        "Milk_Soap"
    ],

    "city":[
        "gurugram",
        "delhi",
        "gurugram",
        "chandigarh",
        "delhi",
        None
    ],

    "Price":[
        150,
        100,
        200,
        180,
        None,
        120
    ],

    "Stock":[
        2000,
        23000,
        14000,
        5000,
        7000,
        None
    ]
}

df = pd.DataFrame(data)
print(df)


sales_data = {

    "Soap_ID":[101,102,103,104,105,106],

    "Sales":[
        500,
        700,
        300,
        450,
        600,
        250
    ],

    "Rating":[
        4.5,
        4.0,
        4.8,
        3.9,
        4.2,
        4.7
    ]
}
sales_df = pd.DataFrame(sales_data)
print(sales_data)
```

```python
#Filtering Data (finding important value)

citydata = df[df["city"] == "gurugram"]
print(citydata)
```

```python
#STRIP() ....USE TO REMOVE unnecessary SPACING

df["Soap_name"]= df["Soap_name"].str.strip()
print(df)
```

```python
#upper().....use to make all alphabets in UPPER CASE

df["Soap_name"]=df["Soap_name"].str.upper()
print(df)
```

```python
#lower()..... use to make all alphabets in LOWER CASE 
df["Soap_name"]=df["Soap_name"].str.lower()
print(df)
```

```python
#USE 2 Methods strip() and lower()
df["Soap_name"]=df["Soap_name"].str.strip().str.lower()
print(df)
```

```python
#Handling Missing Values
#fillna() empty space filling....

df["city"]=df["city"].fillna("Unkown city")
print(df)
```

```python
# Qustion fill the all NaN OR Empty spaces
df[["Price","Stock"]]=df[["Price","Stock"]].fillna(0)
print(df)
```

```python
#Handling Missing Values
#dropna() removes rows containing missing values
# df=df.dropna()
# print(df)
```

```python
#Groupby()....

city_price = df.groupby("city")["Price"].sum()
print(city_price)
```

```python
#Data Sorting

sort_value = df.sort_values(by="Price",ascending=False)
print(sort_value)
```

```python
#IN PYTHON MERGE() --->> IN EXCEL VLOOKUP --->> IN SQL JOINS

final_df = pd.merge(sales_df,df, on="Soap_ID")
print(final_df[["Sales","Rating","Soap_name"]])
```

```python
g_df= pd.merge(sales_df ,df,on="Soap_ID")
print(g_df[g_df["city"]=="gurugram"][["Sales","Rating"]])
```

```python
delhi_df= pd.merge(sales_df,df,on="Soap_ID")
print(delhi_df[delhi_df["city"]=="delhi"]["Sales"].sum())
```

```python
# Find the soap having lowest stock.

lowest_df = pd.merge(sales_df,df,on="Soap_ID")
print(lowest_df["Stock"].min())
```

```python
# Find total sales of gurugram city.

sum_df = pd.merge(sales_df,df,on="Soap_ID")
print(sum_df[sum_df["city"]=="gurugram"]["Sales"].sum())
```

```python
# Sawal 2: Mujhe is table me se sirf un soaps ka data alag karke dikhao jinki Rating 4.2 se zyada hai
result = pd.merge(sales_df,df, on="Soap_ID")
print(result)
```

```python
check_rating = pd.merge(sales_df, df, on="Soap_ID")
print(check_rating[check_rating["Rating"] > 4.2]["Soap_name"])
```

```python
# Data Visualization ----->>>  for making charts ....... >>>> Matplotlib  and Seaborn...
import matplotlib.pyplot as plt

# bar chart ....
plt.bar(check_rating["Soap_name"],check_rating["Sales"])

# title for bar chart 
plt.title("Soap sale Report")

plt.xticks(rotation=25)

# show the bar chat....
plt.show()
```
