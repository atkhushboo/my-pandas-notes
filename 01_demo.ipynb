```python
import pandas as pd

data={
    "Soap_name":[" Neam_Soap ","Lemon_Soap  ", "    Aloavera_Soap"],
    "city":["gurugram",None,"chandigarh"],
    "Price":[150,100,200],
    "Stock":[2000,23000,14000]
}

df=pd.DataFrame(data)
print(df)
```

```python
#STRIP() ....USE TO REMOVE unnecessary SPACING

df["Soap_name"]= df["Soap_name"].str.strip()
print(df)
```

```python
#upper().....use to make all alphabets in UPPER CASE

df["Soap_name"]=df["Soap_name"].str.upper()
print(df)
```

```python
#lower()..... use to make all alphabets in LOWER CASE 
df["Soap_name"]=df["Soap_name"].str.lower()
print(df)
```

```python
#USE 2 Methods strip() and lower()
df["Soap_name"]=df["Soap_name"].str.strip().str.lower()
print(df)
```

```python
#Handling Missing Values
#fillna() empty space filling....

df["city"]=df["city"].fillna("Unkown city")
print(df)
```
